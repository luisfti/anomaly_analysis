```python
# Imports ======================================================================================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
# Data reading =================================================================================================================
dfo = pd.read_csv('/Users/Luis1/Projetos/Detection/datasets/csv/btc_orderbook_df.csv')
dft = pd.read_csv('/Users/Luis1/Projetos/Detection/datasets/csv/btc_trades_df.csv')
dfo.head(5), dft.head(5)
```

```python
# Data standardization =========================================================================================================
scaler = StandardScaler()
dfo[['price_T','volume_T']] = scaler.fit_transform(dfo[['price','volume']])
dft[['price_T','amount_T']] = scaler.fit_transform(dft[['price','amount']])
dfo.head(5), dft.head(5)
```

```python
dbscanO = DBSCAN(eps=0.5, min_samples = 12, metric='euclidean', algorithm='ball_tree', leaf_size=30)
dbscanO.fit(dfo[['price_T','volume_T']])
dfo['dbscan'] = dbscanO.labels_
dfo
```

```python
dbscanT = DBSCAN(eps=0.1, min_samples = 12, metric='euclidean')
dbscanT.fit(dft[['price_T','amount_T']])
dft['dbscan'] = dbscanT.labels_
dft
```

```python
outliers_dfo = dfo[dfo.dbscan == -1]
clusters_dfo = dfo[dfo.dbscan != -1]

colors = dbscanO.labels_
color_clusters = colors[colors != -1]
color_outliers = 'red'

clusters = Counter(dbscanO.labels_)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])
ax.scatter(x=clusters_dfo['price'], y=clusters_dfo['volume'], c=color_clusters, edgecolor='black', s=50)
ax.scatter(x=outliers_dfo['price'], y=outliers_dfo['volume'], c=color_outliers, edgecolor='black',s=50)

ax.set_xlabel('price', family='Arial', fontsize=9)
ax.set_ylabel('volume', family='Arial', fontsize=9)

plt.title('OrderBook Cluster', family='Arial', fontsize=12)
plt.grid(which='major', color='#cccccc', alpha=0.45)
plt.show()

print(clusters)
print(dfo[dfo.dbscan == -1].head())
print('Number of clusters = {}'.format(len(clusters)-1))
```
